```python
import requests
import json
import pandas as pd

# 台灣證券交易所的API URL(上市公司交易資訊)
url = 'https://openapi.twse.com.tw/v1/exchangeReport/STOCK_DAY_ALL' # 日
#url = 'https://openapi.twse.com.tw/v1/exchangeReport/FMSRFK_ALL' # 月
#url = 'https://openapi.twse.com.tw/v1/exchangeReport/FMNPTK_ALL' # 年

# 發送請求
response = requests.get(url)

# 確認回應狀態
if response.status_code == 200:
    # 解析回應的JSON資料
    data = response.json()

    #print(json.dumps(data, indent=4, ensure_ascii=False))
else:
    print(f"Error: {response.status_code}")
```

```python
# 根據股票代碼取得資訊
def find_stock_by_code(code, data):
  for stock in data:
    if stock["Code"] == code:
      return stock
  return None

# 根據公司名稱取得資訊
def find_stock_by_name(name, data):
  for stock in data:
    if stock["Name"] == name:
      return stock
  return None
```

```python
# 測試2330
stock_data = find_stock_by_code('2330', data)
print("根據代碼結果")
pd.DataFrame(stock_data, index=[0])
```

```python
# 測試台積電
stock_data = find_stock_by_name('台積電', data)
print("根據名稱結果}")
pd.DataFrame(stock_data, index=[0])
```

```python
import pandas as pd
def get_stockinfo(soup):
  find_div = soup.find('div', {'class': 'Pos(r)'})
  ul_tag = soup.find('ul', {'class': 'D(f) Fld(c) Flw(w) H(192px) Mx(-16px)'})
  spam_tag = ul_tag.find_all('span')
  result_dic = {}
  filter = [spam_tag[i].text for i in range(len(spam_tag)) if len(spam_tag[i].text) != 0]
  for i in range(len(filter)):
    if i % 2 == 0:
      result_dic[filter[i]] = filter[i+1]

  df = pd.DataFrame(result_dic, index=[0])
  return df
```

```python
import requests
from bs4 import BeautifulSoup

# 設定 Yahoo 奇摩股市的 URL
stock_code = '0050'  # 例如：台積電的股票代碼
url = f"https://tw.stock.yahoo.com/quote/{stock_code}.TW"

# 發送 HTTP GET 請求取得網頁內容
response = requests.get(url)

# 檢查請求是否成功
if response.status_code == 200:
  # 使用 BeautifulSoup 解析 HTML
  soup = BeautifulSoup(response.text, 'html.parser')
else:
  print(f"無法取得資料，HTTP 狀態碼: {response.status_code}")

result = get_stockinfo(soup)
result
```
